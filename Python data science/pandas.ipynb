```python
dataframe_ol = dataframe.copy()
dataframe_ol
```

```python
# Selecionando os nomes das colunas numéricas
columns = dataframe_ol.select_dtypes(include='number').columns

# Obter o percentil
for column in columns:
    lower = round(dataframe_ol[column].quantile(0.01),2)
    upper = round(dataframe_ol[column].quantile(0.995),2)
    print(f'Coluna: {column}\nInferior: {lower}\nSuperior: {upper}')
    print('=='*15)
    dataframe_ol[column] = dataframe_ol[column].clip(lower,upper)
```
